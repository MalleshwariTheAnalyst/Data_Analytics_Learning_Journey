- Let's discuss about three KEYWORDS -- break || Continue || pass
- Break Statement - if you apply break statement in a loop then it will end the loop
- pass - skips the block of code(function , class etc)
- continue - skips one [1] step/iteration during the loop
- Break - Jumps out of the function/loop

```python
for i in range(1 , 11):
    print(i)
```

```python
for i in range(1, 11):
    if i ==5:
        break
    print(i)
```

- Eg ., Let's print the number from 1-20 and we don't want print number which is divisible by 5

```python
nums = [12,15 , 18 , 21,26]
for num in nums:
    if num % 5 == 0:
        print(num)
```

- Prime Number --> How to Check given number is prime number or not!!

```python
num = 12
for i in range(2 , num):
    if num % i ==0:
        print("not prime number")
        break
else:
    print("Prime number")
```
